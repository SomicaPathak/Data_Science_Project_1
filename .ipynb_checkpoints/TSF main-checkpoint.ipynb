```python
#step 1-Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#step 2-Importing data
dt=pd.read_csv("data.csv")
print("Data imported successfully")
```

```python
print("checking the dataset")
dt.head(9)
```

##### Visualising the data with the help of different graphs

```python
dt.plot(x='Hours',y='Scores',style='1')
plt.title('Hours vs Scores')
plt.xlabel('study hours')
plt.ylabel('Percentage score')
plt.show()
```

```python
dt.plot.pie(x='Hours',y='Scores')
```

```python
#step 4-preperaing data 
a=dt.iloc[:,:-1].values
b=dt.iloc[:,1].values
```

```python
#step-6 splitting data for trainning and testing
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test=train_test_split(a,b,train_size=0.8,random_state=0)
```

##### Tranning and testing of model 

```python
#step-6 trainning
re=LinearRegression()
re.fit(a_train,b_train)
print("training done")
```

```python
#step-6 testing
print(a_test)
print("score pridiction")
b_pred=re.predict(a_test)
print(b_pred)
```

##### checking accuracy

```python
df=pd.DataFrame({'Actual':b_test,'Predict':b_pred})
print(df)
```

```python
df.plot(kind='bar')
plt.show()
```

```python
#step 7 using a custom input
hour=[[7.2]]
prid=re.predict(hour)
print(prid)
```

##### Evaluation

```python

print('mean absolute error:',metrics.mean_absolute_error(b_test,b_pred))
print('mean squared error:',metrics.mean_squared_error(b_test,b_pred))
print('Root mean Squared error',np.sqrt(metrics.mean_squared_error(b_test,b_pred)))
```
